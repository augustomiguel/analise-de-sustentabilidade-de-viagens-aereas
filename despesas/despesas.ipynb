```python
from despesas import Despesas
import pandas as pd
import numpy as np
# Cria uma instância da classe Despesas para o ano de 2025 e mês de janeiro
#e salva como dataframe

# Obs: O ano e mês podem ser alterados conforme necessário


desp = Despesas(2025, 1)
despesas = desp.pegarDespesas()
# Converte o dicionário de DataFrames em um único DataFrame
despesas = pd.concat(despesas.values(), ignore_index=True)
```
