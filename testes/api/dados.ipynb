```python
import requests
from datetime import datetime, timedelta
import json

class ConsultaViagens:
    def __init__(self, chave_api):
        self.chave_api = chave_api
        self.base_url = "https://api.portaldatransparencia.gov.br/api-de-dados/viagens"
        self.headers = {
            'chave-api-dados': self.chave_api,
            'Accept': 'application/json'
        }
        self.timeout = 30

    def verificar_conexao_api(self):
        """Verifica se a API está respondendo"""
        try:
            test_params = {
                'dataIdaDe': '2024-01-01',
                'dataIdaAte': '2024-01-02',
                'dataRetornoDe': '2024-01-01',
                'dataRetornoAte': '2024-01-02',
                'pagina': 1
            }
            response = requests.get(
                self.base_url,
                params=test_params,
                headers=self.headers,
                timeout=self.timeout
            )
            return response.status_code == 200
        except:
            return False

    def formatar_data(self, data_str):
        """Converte data de DD/MM/AAAA para YYYY-MM-DD com validação"""
        try:
            data_obj = datetime.strptime(data_str, "%d/%m/%Y")
            return data_obj.strftime("%Y-%m-%d")
        except ValueError:
            raise ValueError(f"Formato de data inválido: {data_str}. Use DD/MM/AAAA")

    def consultar(self, data_ida_de, data_ida_ate, data_retorno_de, data_retorno_ate, **kwargs):
        """Consulta com tratamento completo de erros"""
        params = {
            'dataIdaDe': self.formatar_data(data_ida_de),
            'dataIdaAte': self.formatar_data(data_ida_ate),
            'dataRetornoDe': self.formatar_data(data_retorno_de),
            'dataRetornoAte': self.formatar_data(data_retorno_ate),
            'pagina': kwargs.get('pagina', 1)
        }

        if kwargs.get('codigo_orgao'):
            params['codigoOrgao'] = kwargs['codigo_orgao']

        try:
            response = requests.get(
                self.base_url,
                params=params,
                headers=self.headers,
                timeout=self.timeout
            )

            print(f"\nDEBUG - URL da Requisição:\n{response.url}\n")

            if response.status_code == 200:
                return response.json()
            else:
                error_msg = f"Status {response.status_code}"
                try:
                    error_details = response.json()
                    error_msg += f"\nDetalhes: {json.dumps(error_details, indent=2)}"
                except:
                    error_msg += f"\nResposta: {response.text}"
                raise ValueError(error_msg)

        except requests.exceptions.RequestException as e:
            raise ConnectionError(f"Falha na conexão: {str(e)}")

def main():
    print("\n" + "="*60)
    print(f"{' CONSULTOR DE VIAGENS OFICIAIS ':=^60}")
    print(f"{'='*60}\n")

    # Configuração - SUA CHAVE AQUI (obrigatório)
    CHAVE_API = "9b8e00db8253945fc8e90aa1cd4423be"  # Substitua pela sua chave real

    if not CHAVE_API or len(CHAVE_API) != 32:
        print("ERRO: Chave de API inválida. Deve ter exatamente 32 caracteres.")
        print("Obtenha sua chave em: https://portaldatransparencia.gov.br/api-de-dados/cadastrar-email")
        return

    consultor = ConsultaViagens(CHAVE_API)

    # Verificação inicial da API
    if not consultor.verificar_conexao_api():
        print("ERRO: Não foi possível conectar à API. Verifique:")
        print("- Sua conexão com a internet")
        print("- Se a API está online")
        print("- Sua chave de acesso")
        return

    try:
        # Datas de exemplo (últimos 7 dias)
        data_fim = datetime.now()
        data_inicio = data_fim - timedelta(days=7)

        print(f"Consultando viagens entre {data_inicio.strftime('%d/%m/%Y')} e {data_fim.strftime('%d/%m/%Y')}...")

        resultados = consultor.consultar(
            data_ida_de=data_inicio.strftime("%d/%m/%Y"),
            data_ida_ate=data_fim.strftime("%d/%m/%Y"),
            data_retorno_de=data_inicio.strftime("%d/%m/%Y"),
            data_retorno_ate=data_fim.strftime("%d/%m/%Y"),
            codigo_orgao="00018"  # Ministério da Economia
        )

        if resultados:
            print(f"\nTotal de viagens encontradas: {len(resultados)}")
            for i, viagem in enumerate(resultados[:3], 1):
                print(f"\n[Viagem {i}]")
                print(f"Servidor: {viagem.get('nomeServidor', 'Não informado')}")
                print(f"Destino: {viagem.get('destinos', 'Não informado')}")
                print(f"Período: {viagem.get('dataIda')} a {viagem.get('dataRetorno')}")
                print(f"Valor: R$ {float(viagem.get('valorTotal', 0)):,.2f}")
        else:
            print("Nenhuma viagem encontrada no período especificado.")

    except ValueError as ve:
        print(f"\nERRO DE PARÂMETROS:\n{str(ve)}")
    except ConnectionError as ce:
        print(f"\nERRO DE CONEXÃO:\n{str(ce)}")
    except Exception as e:
        print(f"\nERRO INESPERADO:\n{type(e).__name__}: {str(e)}")
    finally:
        print("\nConsulta finalizada.")

if __name__ == "__main__":
    main()
```
